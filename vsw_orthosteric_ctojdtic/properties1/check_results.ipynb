```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("final_smiles_props_filters_bbb.txt")
df
```

```python
def conditions(s):
    if (s['FILTERS'] == "1") & (s['BBB'] == 1) & ((s['LogS'] >= -5.0) & (s['LogS'] <= -1.0)):
        return 1
    else:
        return 0
```

```python
df['Extract'] = df.apply(conditions, axis=1)
df
```

```python
df1 = df[df["Extract"]==1]
df1
```

```python
df_group = df.groupby("NAMES").count()
df_group[df_group["SMILES"]>1]
```

```python
nums = df.shape[0]
nums
```

```python
index_list = []
for i in range(nums):
    index_list.append(i)
index_list
```

```python
df["INDEX1"] = index_list
```

```python
df
```

```python
df.to_csv("final_smiles_props_filters_bbb_index.txt", index=False)
```
