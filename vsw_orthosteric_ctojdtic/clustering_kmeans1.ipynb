```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
df = pd.read_csv("contact_fp.csv")
df
```

```python
df1 = df.loc[:, (df != 0).any(axis=0)]
df1
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
kmeans1 = KMeans(n_clusters = 2, max_iter=1000)
kmeans1_labels = kmeans1.fit_predict(df1)
```

```python
df2 = pd.DataFrame()
```

```python
df2["Label"]=kmeans1_labels
df2
```

```python
df_full = pd.read_csv("interaction_fingerprint_bymaestro.csv")
df_full
```

```python
df2["Title"] = df_full["Title"]
df2
```

```python
df2.groupby("Label").count()
```

```python
df2.groupby("Title").count()
```

```python
df2.to_csv("name_cluster.csv", index=False)
```
